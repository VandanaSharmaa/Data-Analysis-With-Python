```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
happy= pd.read_excel(r"C:\Users\vansh\OneDrive\Desktop\worldhappinessreport.xlsx")
happy.head()
```

```python
happy.isnull().sum()
```

```python
data= happy.rename({'Explained by: GDP per capita': 'GDP per capita','Explained by: Social support':'Social Support', 'Explained by: Healthy life expectancy':'Healthy life expectancy' }, axis = True)
data
```

```python
D= data.rename({'Explained by: Freedom to make life choices':' Freedom to make life choices','Explained by: Generosity':'Generosity','Explained by: Perceptions of corruption': 'Perceptions of corruption'  }, axis= True)
D
```

```python
happy_d= D.corr()
happy_d
```

```python
D.describe()
```

```python
plt.rcParams['figure.facecolor']= '#FFE5B4'
plt.rcParams['figure.figsize']= (15,7)
sns.heatmap(happy_d, annot=True)
```

```python
plt.rcParams['figure.facecolor']= '#FFE5B4'
plt.rcParams['figure.figsize']= (15,7)
sns.scatterplot(x='Happiness score', y= 'GDP per capita',data= D, hue= D.RANK, s=300)
plt.legend(loc= 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP Per Capita')
plt.title('Happiness VS GDP', fontsize='15')
                                
    
```
